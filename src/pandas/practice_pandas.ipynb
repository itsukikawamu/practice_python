```python
df_person1 = pd.read_csv("personNo_1.csv", index_col = "社員番号")
df_person2 = pd.read_csv("personNo_2.csv", index_col = "社員番号")
```

```python
df_person = df_person1.join(df_person2, how = "outer") #インデックスカラムをキーにする
#デフォルトはleftjoin
```

```python
df_section = pd.read_csv("sectionNo.csv", index_col = "組織番号")
```

```python
pd.merge(df_person, df_section, on=["組織番号", "組織名"], how="left")
#デフォルトはinnerjoin
```

```python
df_person1B = pd.read_csv("personNo_1B.csv", index_col="社員番号")[["氏名"]]
pd.concat([df_person1, df_person1B])
```
