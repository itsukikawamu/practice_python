```python
import pandas as pd
```

```python
train = pd.read_csv("train.tsv", sep="\t", index_col=0)
test = pd.read_csv("test.tsv", sep="\t", index_col=0)
```

```python
data = pd.concat([train, test])
```

```python
data['age'] = data['age'].fillna(data['age'].mean()) 
data['embarked'] = data['embarked'].fillna(data['embarked'].mode()) 
```

```python
data = pd.get_dummies(data)
```

```python
train = data.loc[train.index]
test = data.loc[test.index]
test = test.drop(["survived"], axis=1)
```

```python
train.head()
```

```python
y = train["survived"]
X = train.drop(["survived"], axis=1) 
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X, y)
```

```python
pred = model.predict_proba(test)[:, 1]
test 
```

```python
sample_submit = pd.DataFrame()
sample_submit[0] = test.index
sample_submit[1] = pred
sample_submit.to_csv('submit2.tsv', header=None, sep='\t', index=False)
```
